```python
model_name = "gpt-4o-ca"
```

# 使用 ConversationBufferMemory

```python
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.chains.conversation.memory import ConversationBufferMemory

llm = ChatOpenAI(
    model_name=model_name,
    temperature=0.5,
)

conversation = ConversationChain(llm=llm, memory=ConversationBufferMemory(), verbose=True)

print(conversation.invoke("我姐姐明天要过生日，我需要一束生日花束。"))
print("\n第一次对话后时提示:\n",conversation.prompt.template)
print("\n第一次对话后的记忆:\n", conversation.memory.buffer)

print(conversation.invoke("她喜欢粉色玫瑰，颜色是粉色的。"))
print("\n第二次对话后时提示:\n",conversation.prompt.template)
print("\n第二次对话后的记忆:\n", conversation.memory.buffer)

print(conversation.invoke("我又来了，还记得我昨天为什么要来买花吗？"))
print("\n第三次对话后时提示:\n",conversation.prompt.template)
print("\n第三次对话后的记忆:\n", conversation.memory.buffer)
```

# 使用 ConversationBufferWindowMemory

```python
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.chains.conversation.memory import ConversationBufferWindowMemory

llm = ChatOpenAI(
    model_name=model_name,
    temperature=0.5,
)

conversation = ConversationChain(llm=llm, memory=ConversationBufferWindowMemory(k=1), verbose=True)

print(conversation.invoke("我姐姐明天要过生日，我需要一束生日花束。"))
print("\n第一次对话后的记忆:\n", conversation.memory.buffer)


print(conversation.invoke("她喜欢粉色玫瑰，颜色是粉色的。"))
print("\n第二次对话后的记忆:\n", conversation.memory.buffer)


print(conversation.invoke("我又来了，还记得我昨天为什么要来买花吗？"))
print("\n第三次对话后的记忆:\n", conversation.memory.buffer)
```

# 使用 ConversationSummaryMemory

```python
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.chains.conversation.memory import ConversationSummaryMemory

llm = ChatOpenAI(
    model_name=model_name,
    temperature=0.5,
)

conversation = ConversationChain(llm=llm, memory=ConversationSummaryMemory(llm = llm), verbose=True)

print(conversation.invoke("我姐姐明天要过生日，我需要一束生日花束。"))
print("\n第一次对话后的记忆:\n", conversation.memory.buffer)


print(conversation.invoke("她喜欢粉色玫瑰，颜色是粉色的。"))
print("\n第二次对话后的记忆:\n", conversation.memory.buffer)


print(conversation.invoke("我又来了，还记得我昨天为什么要来买花吗？"))
print("\n第三次对话后的记忆:\n", conversation.memory.buffer)
```

# 使用 ConversationSummaryBufferMemory

```python
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.chains.conversation.memory import ConversationSummaryBufferMemory

llm = ChatOpenAI(
    model_name=model_name,
    temperature=0.5,
)

conversation = ConversationChain(llm=llm, memory=ConversationSummaryBufferMemory(llm = llm, max_token_limit=300), verbose=True)

print(conversation.invoke("我姐姐明天要过生日，我需要一束生日花束。"))
print("\n第一次对话后的记忆:\n", conversation.memory.buffer)


print(conversation.invoke("她喜欢粉色玫瑰，颜色是粉色的。"))
print("\n第二次对话后的记忆:\n", conversation.memory.buffer)


print(conversation.invoke("我又来了，还记得我昨天为什么要来买花吗？"))
print("\n第三次对话后的记忆:\n", conversation.memory.buffer)
```
