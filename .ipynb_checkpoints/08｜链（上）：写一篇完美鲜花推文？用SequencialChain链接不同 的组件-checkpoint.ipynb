# LLMChainï¼šæœ€ç®€å•çš„é“¾

```python
from langchain.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain.chains import LLMChain

template = "{flower}çš„èŠ±è¯­æ˜¯?"

llm = ChatOpenAI(model_name=model_name, temperature=0)
prompt = PromptTemplate(
        template = template,
        input_variables = ["flower"],
    )

llm_chain = prompt | llm

result = llm_chain.invoke("ç™¾åˆ")
print(result)
print(result.content)
```

# Sequential Chainï¼šé¡ºåºé“¾

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SequentialChain

llm = ChatOpenAI(model_name=model_name, temperature=0.7)

template="""
ä½ æ˜¯ä¸€ä¸ªæ¤ç‰©å­¦å®¶ï¼Œç»™å®šèŠ±çš„åç§°å’Œç±»å‹ï¼Œä½ éœ€è¦ä¸ºè¿™ç§èŠ±å†™ä¸€ä¸ª200å­—å·¦å³çš„ä»‹ç»ã€‚

èŠ±å: {name}
é¢œè‰²: {color}
æ¤ç‰©å­¦å®¶: è¿™æ˜¯å…³äºä¸Šè¿°èŠ±çš„ä»‹ç»:"""

prompt_template = PromptTemplate(input_variables=["name", "color"], template=template)

introduction_chain = LLMChain(llm=llm, prompt=prompt_template, output_key="introduction", verbose=True)


template = """
ä½ æ˜¯ä¸€ä½é²œèŠ±è¯„è®ºå®¶ã€‚ç»™å®šä¸€ç§èŠ±çš„ä»‹ç»ï¼Œä½ éœ€è¦ä¸ºè¿™ç§èŠ±å†™ä¸€ç¯‡200å­—å·¦å³çš„è¯„è®ºã€‚
é²œèŠ±ä»‹ç»:
{introduction}
èŠ±è¯„äººå¯¹ä¸Šè¿°èŠ±çš„è¯„è®º:"""
prompt_template = PromptTemplate(input_variables=["introduction"], template=template)

review_chain = LLMChain(llm=llm, prompt=prompt_template, output_key="review", verbose=True)

template = """
ä½ æ˜¯ä¸€å®¶èŠ±åº—çš„ç¤¾äº¤åª’ä½“ç»ç†ã€‚ç»™å®šä¸€ç§èŠ±çš„ä»‹ç»å’Œè¯„è®ºï¼Œä½ éœ€è¦ä¸ºè¿™ç§èŠ±å†™ä¸€ç¯‡ç¤¾äº¤åª’ä½“çš„å¸–å­ï¼Œ300å­—å·¦å³
é²œèŠ±ä»‹ç»:
{introduction}
èŠ±è¯„äººå¯¹ä¸Šè¿°èŠ±çš„è¯„è®º:
{review}
ç¤¾äº¤åª’ä½“å¸–å­:
"""
prompt_template = PromptTemplate(input_variables=["introduction", "review"], template=template)
social_post_chain = LLMChain(llm=llm, prompt=prompt_template, output_key="social_post_text", verbose=True)

overall_chain = SequentialChain(
    chains=[introduction_chain, review_chain, social_post_chain],
    input_variables=["name", "color"],
    verbose=True,
)

result = overall_chain.invoke({"name": "ç™¾åˆ", "color": "ç™½è‰²"})
print(result)
```

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SequentialChain
from langchain.output_parsers import StructuredOutputParser, ResponseSchema

llm = ChatOpenAI(model_name=model_name, temperature=0.7)

response_schemas = [
 ResponseSchema(name="description", description="é²œèŠ±çš„æè¿°æ–‡æ¡ˆ"),
 ResponseSchema(name="reason", description="é—®ä»€ä¹ˆè¦è¿™æ ·å†™è¿™ä¸ªæ–‡æ¡ˆ")
]

output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
print(output_parser.get_format_instructions())

template="""
æ‚¨æ˜¯ä¸€ä½ä¸“ä¸šçš„é²œèŠ±åº—æ–‡æ¡ˆæ’°å†™å‘˜ã€‚
å¯¹äºå”®ä»·ä¸º {price} å…ƒçš„ {flower_name} ï¼Œæ‚¨èƒ½æä¾›ä¸€ä¸ªå¸å¼•äººçš„ç®€çŸ­æè¿°å—ï¼Ÿ
è¾“å‡ºæ ¼å¼:
{format_instructions}"""

prompt_template = PromptTemplate(
    input_variables=["price", "flower_name"], 
    template=template,
    partial_variables={"format_instructions": output_parser.get_format_instructions()},
    output_parser=output_parser,
)

llm_chain = LLMChain(llm=llm, prompt=prompt_template, output_key="introduction", verbose=True)

flowers = ["ç«ç‘°", "ç™¾åˆ", "åº·ä¹ƒé¦¨"]
prices = ["50", "30", "20"]

for flower, price in zip(flowers, prices):
    result = llm_chain.run({"price": price, "flower_name": flower})
    print(result)
```
